```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
dframe= pd.read_csv('data/spatial_autocorrelation.csv')
dframe.columns = dframe.columns.str.strip()
```

For the sake of simplicity, removing the Scaled and Lagged scaled columns.

```python
dframe = dframe.drop(['Walkability Index_Scaled','Walkability Index_Lagged_Scaled'],axis=1)
```

```python
dframe['Diff'] = abs(dframe['Walkability Index'].sub(dframe['Walkability Index_Lagged'],axis=0))
# dframe['Scaled Diff'] = abs(dframe['Walkability Index_Scaled'].sub(dframe['Walkability Index_Lagged_Scaled'],axis=0))
```

The new column developed above has the differences between normal and lagged walkability indices of the SA2 Suburbs. As the lagged walkability index denotes the walkability of neighboring spaces, the differences can tell us how similar or different the area is with its neighbors, in terms of walkability.

```python
dframe.sort_values('Diff',ascending=False)
```
