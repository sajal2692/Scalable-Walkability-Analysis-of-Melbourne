## Correlations with Socio-Economic and Health Metrics 

We finished merging different data from different sources into a single csv file in the previous notebook. Here we'll use the csv to check out if these _features_ for Inner Melbourne Suburbs are interrelated or not. 

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import seaborn as sns
from scipy import stats

%matplotlib inline
```

```python
dframe = pd.read_csv('data/sa2_cor_features.csv') #Reading the new csv file
dframe.columns = dframe.columns.str.strip()
dframe.drop('Unnamed: 0',axis=1,inplace=True)
dframe.set_index('area_name',inplace=True) #Setting the index to SA2 name
```

Here is what the datafreame looks like. 

```python
dframe
```

Pandas allows us to build a correlation matrix for all the columns in our dataframe!

```python
dframe.corr(method='pearson') #Using Pearson's coefficient of correlation
```

```python
r_r2 = DataFrame(dframe.corr(method='pearson')['Walkability Index'])
```

Let's build another column in our dataset, for "Coefficient of Determination'.

```python
r_r2.columns = ['Correlation'] 
r_r2['Coefficient of Determination'] = r_r2['Correlation'] ** 2
```

```python
r_r2
```

Now, we can visualise the correlation between the Walability of a suburb, and a related socio-economic index using scatter plots with the line of best fit. 

For the sake of this analysis, we assume a threshold of 0.3 to determine a positive or negative correlation. Anything above 0.3 will be positively correlated, and below -0.3 negatively correlated. Correlations between the threshold are deemed inconsequential.

Lets start with Health Index: Obesity and Overweight %. 

```python
#For Health Indices
for feature in dframe.columns[1:3]:
    scatter_plot = sns.lmplot('Walkability Index',feature,data=dframe) #Creating inline figures
    scatter_plot.savefig("figures/"+feature+'_walk_corr.png') #Saving figures
    
```

#### Health Index: Obesity and Overweight %

Both the obesity and overweight percentages seem to be negatively correlated with the walkability index, so there is indeed some degree of truth to the fact that suburbs with a higher walkability score have a lower percentage of overweight and obese persons living there. 

By overweight, we mean that a persons weight is maybe slightly above a healthy weight range; being obese, on the other hand, is a severely unhealthy condition. We see that theres a much tighter negative correlation of walkability with the obesity rate, than the overweight rate. People might walk to school, work or shopping in more walkable areas, but they still can be a little overweight as walking is a basic physical activity. But, obesity indeed seems to be prevented by doing a little walking.

```python
#For public transport stop and station counts
for feature in dframe.columns[9:]:
    scatter_plot = sns.lmplot('Walkability Index',feature,data=dframe) #Creating inline figures
    scatter_plot.savefig("figures/"+feature+'_walk_corr.png') #Saving figures
```

#### Public Transport Stop and Station Counts

In its original form, the data-sets contained lists of bus stops, tram stops, train stations, and their respective coordinates. A point in polygon join was done to assign them their respective SA2 Suburbs. Then they were aggregated on the basis of these suburbs using pandas. We perform correlation on the counts of bus and tram stops, and the train stations in an SA2 suburb, with the walkability index. 

In case of bus and tram stops, there seems to be a mixed conclusion. More walkable areas seem to have less bus stops, but more tram stops. While in the case of the number of train stations, theres almost no correlation at all. 

It might be safe to assume that even though the counts of public transport points seem to be somewhat related to the walkability scores (at least in the case of bus and tram stops), whatever correlation there might be, is not directly dependent on the walkability score. Its more dependent on the type of area, and how the public transportation system of Melbourne is planned out. Cities have fewer bus stops, with not many bus routes going through cities. Melbourne has trams for that. While suburbs with little to no tram connectivity make up for it with the bus connectivity.
