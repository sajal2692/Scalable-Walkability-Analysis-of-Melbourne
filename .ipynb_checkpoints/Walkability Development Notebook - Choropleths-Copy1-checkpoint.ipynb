```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
dframe = pd.read_csv('dev_data.csv')
dframe.columns = dframe.columns.str.strip()
```

```python
#dframe
```

```python
#Dropping columns deemed irrelevant
aframe = dframe.drop(['UID','sa3_code11','sa4_code11','sa2_main11','sa3_name11','sa4_name11','mb_code11'],axis=1)
#maybe not delete mb_code11 and/or sa1_main11
```

```python
aframe
```

```python
#Group by suburb
avg_sa2 = aframe[['sa2_name11','SumZScore','Connectivity_ZScore','AverageDensity_ZScore','LandUseMixMeasure_ZScore']].groupby('sa2_name11').mean()
```

```python
avg_sa2
```

```python
avg_sa2.idxmin()
```

```python
avg_sa2.idxmax()
```

```python
#import sys
#sys.path.insert(0,'..')
import folium
print (folium.__file__)
print (folium.__version__)
```

```python
suburb_walk_avg = avg_sa2[['SumZScore']].reset_index()
suburb_walk_avg.columns = ['SA2 Suburb','Walkability Score']
suburb_walk_avg
```

```python
map_melb_sa2 = folium.Map(location=[-37.814,144.954],zoom_start = 13,max_zoom=15)

map_melb_sa2.choropleth(
    geo_path='sa2boundaries.json',
    data = suburb_walk_avg,
    key_on= 'properties.area_name',
    columns=['SA2 Suburb', 'Walkability Score'],
    fill_color='YlGn',
    line_weight=2,
    legend_name = 'Walkability Score'
)

map_melb_sa2
```
